```python
# Passo a Passo:
    # [PASSO 00] Entender o Desafio da Empresa;
    # [PASSO 01] Importar a base de dados;
import pandas as pd

table = pd.read_csv('./gabarito/clientes.csv')
table = table.drop(columns='id_cliente')
display(table)
display(table.info())
```

```python
    # [PASSO 02] Preparar a base de dados para a IA;
        #converter os textos(objects) para numeros int ou float
            #profissao
            #mix_credito
            #comportamento_pagamento
        #Label Enconder:
'''Ex.:
    mecanico - 1
    medico - 2
    musico - 3
    cientista - 4 '''   
    
from sklearn.preprocessing import LabelEncoder

#profissao
encoder1 = LabelEncoder() 
table['profissao'] = encoder1.fit_transform(table['profissao'])

#mix_credito
encoder2 = LabelEncoder() 
table['mix_credito'] = encoder2.fit_transform(table['mix_credito'])
#comportamento_pagamento
encoder3 = LabelEncoder() 
table['comportamento_pagamento'] = encoder3.fit_transform(table['comportamento_pagamento'])


display(table.info())
display(table)
```

```python
    # [PASSO 03] Criar um modelo de IA -> Nota/Score;
        #é necessário dividir a tabela em X(dados de aprendizado) e Y(dados para previsão);
        # x => num_contas\num_cartoes\investimento_mensal\...
        # y => score_credito
y = table['score_credito']
x = table.drop(columns='score_credito')
    #dividir x e y em treino e teste => x_treino e y_treino \ x_teste e y_teste
    
from sklearn.model_selection import train_test_split
x_train,x_test,y_train, y_test = train_test_split(x,y)



#1. Importar a IA
#2. Criar a IA
#3. Treinar a IA

# 2 Modelos
    # Árvore de decisões

    # Nearest Neighboors
    # [PASSO 03] Criar um modelo de IA -> Nota/Score;
        #é necessário dividir a tabela em X(dados de aprendizado) e Y(dados para previsão);
        # x => num_contas\num_cartoes\investimento_mensal\...
        # y => score_credito
y = table['score_credito']
x = table.drop(columns='score_credito')
    #dividir x e y em treino e teste => x_treino e y_treino \ x_teste e y_teste
    
from sklearn.model_selection import train_test_split
x_train,x_test,y_train, y_test = train_test_split(x,y)



#1. Importar a IA
#2. Criar a IA
#3. Treinar a IA

# 2 Modelos
    # Árvore de decisões -> Random Forest
from sklearn.ensemble import RandomForestClassifier #import IA
model_forest = RandomForestClassifier()#create
model_forest.fit(x_train,y_train)

    # Nearest Neighbors -> KNN
from sklearn.neighbors import KNeighborsClassifier    
model_neighbor = KNeighborsClassifier()
model_neighbor.fit(x_train,y_train)
    
    
```

```python
    # [PASSO 04] Avaliar melhor modelo de IA ;
prev_forest = model_forest.predict(x_test) #testar o modelo para ver qual tera melhor desempenho nas previsões
prev_knn = model_neighbor.predict(x_test)

#accuracy score
from sklearn.metrics import accuracy_score
acc_forest = accuracy_score(y_test,prev_forest)
acc_knn = accuracy_score(y_test,prev_knn)
display(f"Accurary RandomForest = {acc_forest}")
display(f"Accuracy NearestNeighbors = {acc_knn}")
#modelo randomforest demonstrou maior accuracy em relação ao modelo knn
```
